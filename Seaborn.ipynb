```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy.stats import norm
```

```python
df=pd.read_csv('googlePlayStore.csv')
df.head()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(data=df,x='Category')
```

```python
plt.figure(figsize=(12,8))
sns.countplot(data=df,y='Category')
```

```python
df=df.drop(df.loc[df.loc[:,'Type']=='0',:].index)
df.loc[:,'Type'].unique()
```

```python
df=df.drop(df.loc[df.loc[:,'Type']=='0',:].index)
df=df.drop(df.loc[df.loc[:,'Type'].isnull(),:].index)
```

```python
df.loc[:,'Type'].unique()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(data=df,y='Category',hue='Type')
```

```python
plt.figure(figsize=(12,8))
sns.countplot(data=df,y='Category',hue='Type',color='orange')
```

```python
plt.figure(figsize=(12,8))
sns.countplot(data=df,y='Category',hue='Type',palette='Blues_d')
plt.legend(loc='upper right')
plt.title('seaborn_test')
```

```python
f, axes=plt.subplots(2,1,figsize=(20,40))
sns.countplot(data=df,y='Category',hue='Type',palette='Blues_d',ax=axes[0])
sns.boxplot(data=df,y='Category',x='Rating',hue='Type',whis=3,width=0.5,fliersize=1,linewidth=1,ax=axes[1])
axes[0].set_title('seaborn_test')
#plt.legend(loc='upper right')
```

```python
sns.distplot(df['Rating'],bins=50,kde=False,hist=True,fit=norm)
plt.show()
```

```python
f, axes=plt.subplots(2,1,figsize=(12,8))
sns.distplot(df['Rating'],bins=50,kde=False,hist=True,fit=norm,vertical=True,ax=axes[0])
sns.distplot(df['Rating'],bins=50,kde=False,hist=True,fit=norm,vertical=False,ax=axes[1])
plt.show()
```

```python
plt.figure(figsize=(10,18))
sns.boxplot(data=df,y='Category',x='Rating',hue='Type',whis=3,width=0.5,fliersize=1,linewidth=1)
plt.show()
```

```python
plt.figure(figsize=(10,20))
sns.stripplot(data=df,y='Category',x='Rating',jitter=False)
plt.show()
```

```python
##sns.distplot(df['Rating'],hist=False)
plt.figure(figsize=(10,20))
sns.violinplot(data=df,y='Category',x='Rating',hue="Type",split=True,linewidth=1,inner='quartile')
plt.show()
```

```python
sns.lmplot(data=df,x='Rating',y='Installs',hue="Type",col='Type',legend=True)
plt.show()
```
